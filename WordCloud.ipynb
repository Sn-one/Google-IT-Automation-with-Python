```python
with open('project .txt') as file_object:
  contents = file_object.read()
print (contents)
```

```python
def calculate_frequencies(contents):
    # Here is a list of punctuations and uninteresting words you can use to process your text
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    uninteresting_words = ["the", "a", "to", "if", "is", "it", "of", "and", "or", "an", "as", "i", "me", "my", \
    "we", "our", "ours", "you", "your", "yours", "he", "she", "him", "his", "her", "hers", "its", "they", "them", \
    "their", "what", "which", "who", "whom", "this", "that", "am", "are", "was", "were", "be", "been", "being", \
    "have", "has", "had", "do", "does", "did", "but", "at", "by", "with", "from", "here", "when", "where", "how", \
    "all", "any", "both", "each", "few", "more", "some", "such", "no", "nor", "too", "very", "can", "will", "just"]
    
    # LEARNER CODE START HERE
    dict_word = {}
    split_words = contents.split()
    for word in split_words:
        if word in uninteresting_words:
            pass
        else:
            for letter in word:
                if letter in punctuations:
                    letter.replace(punctuations,"")
            if word not in dict_word.keys():
                dict_word[word]=0
            else:
                dict_word[word]+=1
      
    return dict_word
```

```python
word_fq = calculate_frequencies(contents)
```

```python
print (word_fq)
```

```python
import wordcloud
import numpy as np
from matplotlib import pyplot as plt
```

```python
def create_wc_array(dict_word):
  #wordcloud
    cloud = wordcloud.WordCloud()
    cloud.generate_from_frequencies(dict_word)
    return cloud.to_array()
```

```python
myimage = create_wc_array(word_fq)
plt.imshow(myimage, interpolation = 'nearest')
plt.axis('off')
plt.show()
```
